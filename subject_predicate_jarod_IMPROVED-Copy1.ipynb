```python
csvh = open('s_p_order_jarodtest.csv', "w")
header = ['book', 'order']
csvh.write('{}\n'.format(','.join(header)))
for cl in clause_list:
    line = clause_dict[cl]
    csvh.write('{}\n'.format(','.join(line)))
    
csvh.close()
```
